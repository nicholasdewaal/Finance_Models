```python
import time
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict
from IPython.display import clear_output
all_data = defaultdict(list)
```

```python
def logistic(x, r):
    return r * x * (1 - x)

def burn_in(x, r):
    n = 0
    while n < 250:
        x = logistic(x, r)
        n = n + 1
    return x
```

```python
x, r = .5, 3.6 # try r = 2.8 through 3.8 by .05 inc.
history = list()
x = burn_in(x, r)
print(x)
```

```python
n = 0
while n < 40:
    x = logistic(x, r)
    history.append(x)
    clear_output(wait=True)
    print("current value: ", x)
    print("history: ", [round(x,3) for x in history])
    n = n + 1
    time.sleep(.1)
all_data['x'].extend(history)
all_data['r'].extend([r] * len(history))
```

```python
plt.scatter(all_data['r'], all_data['x'], color='blue')
```

```python
def plot_logistic(n, lower, upper, inc, hg):
    h, hr = list(), list()
    for r in np.arange(lower, upper, inc):
        x = burn_in(.5, r)
        for _ in range(n):
            x = logistic(x, r)
            h.append(x)
            hr.append(r)
    plt.figure(figsize=(100,hg))
    plt.scatter(hr, h)
```

```python
plot_logistic(n=30, lower=2.6, upper=3.8, inc=.0001, hg=50)
```
